```python
BASE_DIR = os.getcwd()
BASE_DIR = BASE_DIR.replace(os.sep, '/') + '/Dataset'

# Calling the string columnm numbers as a string;
numbersDataFrame = pandas.read_csv(f'{BASE_DIR}/numbers.csv', dtype={'label' : str})
```

```python
# Exploring the provided dataset;
# Creating a figure that holds the axes for the representation of data;
figure, axis = plot.subplots(nrows=1, ncols=3, figsize=(24, 4))

def GroupData(group, axis):
    numbersDataFrame.groupby([group]).count().rename(columns={'file' : group})[group].plot.bar(ax = axis, subplots=True)

#Organizing the data based on origin, group, and label by axis;
GroupData('origin', axis[0])
GroupData('group', axis[1])
GroupData('label', axis[2])

plot.show()

# Get the first data entries in a data set;
numbersDataFrame.head()
```

```python
#Returning a file item filtered by the group and label parameters input;
def NumberSample(dataFrame, group, number):
    return dataFrame[(dataFrame['group'] == group) & (dataFrame['label'] == number)].iloc[0]['file']

#Takes a sample from the image sampled and plots it;
def SampleAxisPlot(axis, dataFrame, group, number):
    image = NumberSample(dataFrame, group, number)
    axis.imshow(plot.imread(f'{BASE_DIR}/numbers/{image}'))
    axis.title.set_text(group)

#Setting the subplots and positions for the samples;
def PlotGroupSample(dataFrame, groups, number):
    figure, axis = plot.subplots(nrows=1, ncols=len(groups), figsize=(24, 4))

    #For all groups, plot one sample of each group;
    for index, group in enumerate(groups):
        SampleAxisPlot(axis[index], dataFrame, group, number)


groups = ['Hnd', 'Fnt', 'GoodImg', 'BadImag']
PlotGroupSample(numbersDataFrame, groups, '8')
```
